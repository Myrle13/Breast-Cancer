```python
import pandas as pd
import numpy as np
```

```python
# Read the dataset
data = pd.read_csv(r"C:\Users\Myrtle Mai Nguyen\OneDrive\Desktop\Breast Cancer\Breast_Cancer.csv")
data. head()
```

```python
# Drop colomns unnecessary to analysis
df = data.filter(['Age', 'Race','Marital Status','6th Stage', 'Survival Months', 'Status'])
df
```

```python
# Check empty value in dataset
df = df.replace('', np.nan)             # remove empty values
nan_values = df[df.isna().any(axis=1)]  # to get all rows with Na
```

```python
# Check the min and max age in the data
df['Age'].agg(['min', 'max'])
```

```python
# Group age range 
# 30-39, 40-49,50-59,60-69
df['Age'] = pd.cut(df['Age'], bins = [29, 39, 49, 59, float('Inf')], labels = ['30+', '40+','50+', '60+'])
df
```
